```python
import pandas as pd
```

```python
df=pd.read_csv('/content/bank-additional-full organised.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#drop unwanted columns
df=df.drop(['emp.var.rate','cons.price.idx','cons.conf.idx',
            'euribor3m','nr.employed'],axis=1)
```

```python
df.info()
```

```python
#check for duplicates
df.duplicated().sum()
```

```python
#view all duplicate rows
df[df.duplicated()]
```

```python
#remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
#check for missing values
print(df.isnull().sum())
```

```python
#keeping only rows with age between 18 and 90
df = df[(df['age'] >= 18) & (df['age'] <= 90)]
```

```python
#checking new min and max age
print(df['age'].min(), df['age'].max())
```

```python
df.info()
```

```python
#reset index and add id column
df.reset_index(drop=True, inplace=True)
df.insert(0, 'id', range(1, len(df) + 1))
```

```python
df.info()
```

```python
#save the cleaned dataset
df.to_csv("bank_data_clean.csv", index=False)
```
